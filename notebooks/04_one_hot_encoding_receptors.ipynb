For experiments with other models, we need the receptor classes to be one-hot encoded. This is a quick notebook that takes the prepared smiles_to_receptor.csv dataset from make_dataset.py and one-hot encodes it, saving it as smiles_to_receptor_one_hot.csv

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
data = pd.read_csv("../data/processed/smiles_to_receptor.csv")
```

```python
data = data[["canonical_smiles", "pref_name"]]
print(data["pref_name"].nunique())
data.head()
```

```python
y = pd.get_dummies(data.pref_name)
print(y.shape)

y.insert(loc=0, column="canonical_smiles", value=data["canonical_smiles"])
y.head()
```

```python
y.to_csv("../data/processed/smiles_to_receptor_one_hot.csv", index=False)
```
